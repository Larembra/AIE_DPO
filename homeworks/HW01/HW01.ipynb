```python
a = np.zeros((3,3))
print(a)
b = np.ones((3,3))
print(b)
c = np.arange(1,10).reshape((3,3))
print(c)
print(c.shape, c.dtype)
```

```python
print(c[(1,1)])
print(c[0])
```

```python
print(b+c)
print((2*b)*c)

print(c.mean(axis=0)) #среднее значение вдоль столбцов
print(c.mean(axis=1)) #среднее значение вдоль строк

c_min = c.min()
c_max = c.max()


c_normalized = (c - c_min) / (c_max - c_min)

print(c_normalized)
```
